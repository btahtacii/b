This part should be top of the user interface and it should contain two buttons and one checkbox.

***New User Button***
This button should create a new user, when new user button clicked user can add a new user to the database with given columns of database.

***Save User Button***
This button saves credentials of given user. When user clicked that button, the data should be in also database.


***Hide Disabled User Button***
This checkbox hides the users, whose not enabled the checkbox

3.PNG

This part should be located on the right side of user interface. This part is for adding a new user to the database with given input fields.


**Username**
This input field takes a input for assigning to username of the user.

**Display Name**
This input field takes a input for assigning to display name of the user.

**Phone**
This input field takes a  input for assigning to phone of the user.

**Email**
This input field takes a input for assigning to email of the user.

**User Roles**
This input field takes a  input for UserRoles of the user.There options should Guest,Admin,Superadmin

**Enabled**
This input field takes a checkbox input for assigning the value Enabled or disable.
